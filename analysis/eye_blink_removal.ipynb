```python
import pylab as pl
from pylab import arange
from math import e
from math import pi
from math import sin
from math import cos
from numpy import poly1d

sys.path.append("/home/ubuntu/github/bsr/analysis/")
from bsrAnalysisLib import *
#from tasks import *


# need some zero vectors...
def zeroV(m):
    z = [0]*m
    return(z)

#INPUT: n; x0, x1, ... ,xn; a0 = f(x0), a1 =f(x1), ... , an = f(xn).
def cubic_spline(xn,a,xEval):
    """function cubic_spline(xn,a) interpolates between the knots
    specified by lists x and y. The function computes the coefficients
    and outputs the ranges of the piecewise cubic splines."""
    
    n = len(xn)
    
    h = zeroV(n-1)

    # alpha will be values in a system of eq's that will allow us to solve for c
    # and then from there we can find b, d through substitution.
    alpha = zeroV(n-1)

    # l, u, z are used in the method for solving the linear system
    l = zeroV(n+1)
    u = zeroV(n)
    z = zeroV(n+1)

    # b, c, d will be the coefficients along with a.
    b = zeroV(n)     
    c = zeroV(n+1)
    d = zeroV(n)    

    for i in range(n-1):
        # h[i] is used to satisfy the condition that 
        # Si+1(xi+l) = Si(xi+l) for each i = 0,..,n-1
        # i.e., the values at the knots are "doubled up"
        h[i] = xn[i+1]-xn[i]
    
    for i in range(1, n-1):
        # Sets up the linear system and allows us to find c.  Once we have 
        # c then b and d follow in terms of it.
        alpha[i] = (3./h[i])*(a[i+1]-a[i])-(3./h[i-1])*(a[i] - a[i-1])

    # I, II, (part of) III Sets up and solves tridiagonal linear system...
    # I   
    l[0] = 1      
    u[0] = 0      
    z[0] = 0

    # II
    for i in range(1, n-1):
        l[i] = 2*(xn[i+1] - xn[i-1]) - h[i-1]*u[i-1]
        u[i] = h[i]/l[i]
        z[i] = (alpha[i] - h[i-1]*z[i-1])/l[i]

    l[n] = 1
    z[n] = 0
    c[n] = 0

    # III... also find b, d in terms of c.
    for j in range(n-2, -1, -1):      
        c[j] = z[j] - u[j]*c[j+1]
        b[j] = (a[j+1] - a[j])/h[j] - h[j]*(c[j+1] + 2*c[j])/3.
        d[j] = (c[j+1] - c[j])/(3*h[j]) 

    # Now that we have the coefficients it's just a matter of constructing
    # the appropriate polynomials and graphing.

    X = []
    Y = []
    
    for j in range(n-1):
        poly = buildPoly(a[j],b[j],c[j],d[j],xn[j])
        #pts = np.linspace(xn[j],xn[j+1],10)
        
        if j==0:
            cond = (xEval < xn[j+1])
        elif j==n-2:
            cond = (xEval >= xn[j])
        else:
            cond = (xEval >= xn[j])*(xEval < xn[j+1])
        pts = xEval[cond]
        #print xEval[c]
        #print pts
        #print len(pts),len(poly(pts))
        X = np.append(X,pts)
        Y = np.append(Y,poly(pts))
    return X,Y

def buildPoly(a,b,c,d,x_i):
    '''build polynomial function'''
    root = poly1d(x_i,True)
    poly = float(d)*(root)**3 + float(c)*(root)**2 + float(b)*root + float(a)
    return poly  


#def cub_graph(a,b,c,d, x_i, x_i_1):
#    """cub_graph takes the i'th coefficient set along with the x[i] and x[i+1]'th
#       data pts, and constructs the polynomial spline between the two data pts using
#       the poly1d python object (which simply returns a polynomial with a given root."""
#    poly = buildPoly(a,b,c,d,x_i)    
    # Set up our domain between data points, and plot the function
#    pts = arange(x_i,x_i_1, 0.001)
#    pl.plot(pts, poly(pts), '-')
#    return
```

```python
def computeAverageSpline(x,y,xEval):
    
    data = y.copy()
    cMinMax = diff(sign(diff(data))).nonzero()[0] + 1 # local min+max
    cMin = (diff(sign(diff(data))) > 0).nonzero()[0] + 1 # local min
    cMax = (diff(sign(diff(data))) < 0).nonzero()[0] + 1 # local max
    
    cMin = np.concatenate([[0],cMin,[-1]])
    cMax = np.concatenate([[0],cMax,[-1]])
    
    #diff = np.diff(y)
    #cMax = np.r_[True, diff[1:] < diff[:-1]]
    #cMin = np.r_[True, diff[1:] > diff[:-1]]
    #cMax[-1] = True
    #cMin[-1] = True
    
    ySpline = cubic_spline(x,y,xEval)[1]
    yDownSpline = cubic_spline(x[cMin],y[cMin],xEval)[1]
    yUpSpline = cubic_spline(x[cMax],y[cMax],xEval)[1]
    avgSpline = (yUpSpline + yDownSpline)/2.
    h = ySpline - avgSpline
    #return {'avgSpline':avgSpline,'h':h,'yDownSpline':yDownSpline,'yUpSpline':yUpSpline}
    return avgSpline,h

def siftingLoop(x,y,xEval, tol=0.3):
    i = 0
    SD  = 100
    
    #xEval = np.linspace(x[0],x[-1],nbins)
    avgSpline,h_old = computeAverageSpline(x,y,xEval)
    ySpline = avgSpline + h_old
    
    while SD > tol:
        avgSpline,h = computeAverageSpline(xEval,h_old,xEval) 
        SD = np.sum((np.abs(h[1:-1] - h_old[1:-1]))**2/(h_old[1:-1]+0.0000001)**2)
        #print i,SD
        i+=1
        h_old = h.copy()
    return h,ySpline
        

def imfLoop(x,y,n=2,nbins=200):
    print 0
    xEval = np.linspace(x[0],x[-1],nbins)
    c,ySpline = siftingLoop(x,y,xEval)
    r = ySpline - c
    
    C = c
    R = r
    
    pl.figure(1,(10,7))
    pl.plot(c)
        
    pl.figure(2,(10,7))
    pl.plot(r)  
    
    dC = 10000
    #while np.abs(dC) > 0.01:
    for i in range(1,n+1):
        print i
        c = siftingLoop(xEval,r,xEval)[0]
        r = r - c
        
        pl.figure(1,(10,7))
        pl.plot(c)
        
        pl.figure(2,(10,7))
        pl.plot(r)        
        
        C = vstack((C,c))
        R = vstack((R,r))
        dC = np.sum((np.abs(R[-1]**2 - R[-2]**2))**2/R[-2]**2**2)
            
        print dC
    return C,R
```
